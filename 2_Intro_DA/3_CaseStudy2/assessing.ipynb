```python
# number of samples in each dataset
print('Number of samples in 2008 dataset:', df_08.shape[0])
print('Number of samples in 2018 dataset:', df_18.shape[0])
```

```python
# number of columns in each dataset
print('Number of columns in 2008 dataset:', df_08.shape[1])
print('Number of columns in 2018 dataset:', df_18.shape[1])
```

```python
# number of duplicate rows in each dataset
print('Number of duplicate rows in 2008 dataset:', df_08.duplicated().sum())
print('Number of duplicate rows in 2018 dataset:', df_18.duplicated().sum())
```

```python
# datatypes of columns
print('2008 Datatypes:')
print(df_08.dtypes, '\n')
print('2018 Datatypes:')
print(df_18.dtypes)
```

```python
# features with missing values
print('Missing Values in 2008 Dataset:')
print(df_08.isnull().sum(), '\n')
print('Missing Values in 2018 Dataset:')
print(df_18.isnull().sum())
```

```python
# number of non-null unique values for features in each dataset
print('Number of Unique Values in 2008 Dataset:', df_08.nunique().sum())
print('Number of Unique Values in 2018 Dataset:', df_18.nunique().sum())
```

```python
# what those unique values are and counts for each
col_name = list(df_08)

print('Number of Unique Values in 2008 Dataset per Feature:')
for column in df_08:
    print(column + ': '+ str(df_08[column].nunique()))

print('\nNumber of Unique Values in 2018 Dataset per Feature:')
for column in df_18:
    print(column + ': '+ str(df_18[column].nunique()))    
```
