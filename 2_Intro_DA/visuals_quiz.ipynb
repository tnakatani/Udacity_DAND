```python
# imports and load data
import pandas as pd

# This allows us to view visualizations in Jupyter notebook - useful!!
% matplotlib inline

# View summary of census income data
df = pd.read_csv('powerplant_data_edited.csv')
df.info()
```

```python
# plot relationship between temperature and electrical output
df.plot(x='Temperature', y='Net Energy Output', kind='scatter');
```

```python
# plot distribution of humidity
df['Relative Humidity'].hist();
```

```python
# plot box plots for each variable
df.plot(kind='box', figsize=(10,10));
```
